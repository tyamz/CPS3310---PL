```python
## Thomas Yamakaitis
## Project 4
## CPS3310*02 - Programming Languages
## Dr. Morreale
import sys
dictionary = ["public","class","static","void","main","String","args","for","if","while"]
methods = ["System.out.println"]
symbols = ["\n","\t"," ","(",")","[","]",";"]
strings = []
classes = []
reserved = []
calls = []
error = ""
err_count = 0
file = ""
unmatched = 0
matched = 0
line_count = 0

# Main method - scans in file
def main():
    global file
    # File must be in same directory
    f = open("welcome.java", "r")
    file = f.read()
    lex()

# Lexical Analyzer
def lex():
    global error
    global err_count
    global unmatched
    global matched
    i = 0
    c = False
    #print(str(len(file)))
    while i < len(file):
        lexeme = nextLexeme(i)
        #print(lexeme)
        if lexeme == "":
            next
        elif "”" in lexeme:
            strings.append(lexeme)
        elif "{" in lexeme:
            unmatched += 1
        elif "}" in lexeme:
            if unmatched > 0:
                unmatched -= 1
                matched += 1
            else:
                error += "Error: extra \"}\" found... \n"
                #print("Error: extra \"}\" found... \n")
        elif c:
            classes.append(lexeme)
            c = False
        elif lexeme in methods:
            calls.append(lexeme)
        elif lexeme not in dictionary:
            err_count += 1
            error += "Error at line " + str(line_count) + ": \"" + lexeme + "\" not found. \n"
            # print("Error: \"" + lexeme + "\" not found. \n")
        elif lexeme not in reserved:
            if lexeme == "class":
                c = True
            reserved.append(lexeme)
        i += len(lexeme) + 1
    err_count += unmatched
    if err_count > 0:
        print("Errors found: " + str(err_count) + "\n" + error + "\n")
        if unmatched > 0:
            print("Unmatched curly bracket(s): " + str(unmatched))
    else:
        print("Reserved word list: \n" + str(reserved) + "\n")
        print("Class name(s): \n" + str(classes) + "\n")
        print(str(len(strings)) + " variable string(s): \n" +  str(strings) + "\n")
        print("Method calls: \n" + str(calls) + "\n")
        print("Curly brace pair count: " + str(matched))

# Find next lexeme
def nextLexeme(start):
    global file
    global line_count
    lexeme = ""
    i = start
    # print(str(i) + " " + str(len(file)))
    if file[i-1] == "\n":
        line_count += 1
    while (file[i] not in symbols):
        if file[i] == "“":
            while file[i] != "”":
                lexeme += file[i]
                i += 1
            lexeme += file[i]
            return lexeme
        # print(file[i])
        lexeme += file[i]
        i += 1
        if i >= len(file):
            break
    return lexeme
main()
```
